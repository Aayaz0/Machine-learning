```python
#linear search
def linear_search(arr, target):
  for i in range(len(arr)):
    if arr[i] == target:
      return i
  return -1


result=linear_search([40,50,55,57,60,70], 60)
if result == -1:
  print("Element  not found")
else:
   print(f"Element found at index {result}")
```

```python
# Decorators : used to decorate the function
def my_decorator(func):
    print('****')
    func()
    print('*********')

@my_decorator
def hello():
    print('hello there')
hello

@my_decorator
def Bye():
    print('Good bye')
```

```python
count = 0
def perform_transaction():
    global count
    count += 1
    return count
print(perform_transaction())
print(perform_transaction())
print(perform_transaction())
```

```python
# non-local
def outer():
    x = 'this is the outer value of x'
    def inner():
        x ='this is inner value of x'
        print(x)
    inner()
    print(x)
outer()
```

```python
def outer():
    x = 'this is the outer value of x'
    def inner():
        nonlocal x
        x ='this is inner value of x'
        print(x)
    inner()
    print(x)
outer()
```

```python
# wap to implement our prev weight converter calculator make the program continuous and include error handling as well
while True:
    try:
        weight = float(input("What's your weight? "))
        unit = input("Enter the unit (kg/pound): ").lower()

        if unit == 'kg':
            converted = weight / 0.45
            print(f'You are {converted:.2f} pounds.')
            if converted == 100:
                print("Century!")

        elif unit == 'pound':
            converted = weight * 0.45
            print(f'You are {converted:.2f} kg.')
            if converted == 100:
                print("Century!")
        else:
            print("Invalid unit. Please enter 'kg' or 'pound'.")
            break

    except ValueError:
        print("Invalid input")


```

```python
# OrderedDict 
from collections import OrderedDict
d1 = OrderedDict({'a':1,'b':2})
d2 = OrderedDict({'b':1,'a':2})
print(d1==d2)
```

```python
#next way 
from collections import OrderedDict
d1 = OrderedDict({'a':1,'b':2})
d2 = OrderedDict({'b':1,'a':2})
d1['a'] = 1
d1['b'] = 2
d2['b'] = 2
d2['a'] = 1

print(d1==d2)
```

```python
from collections import defaultdict

d1 = defaultdict(int, {'a': 4, 'b': 5})
print(d1['x'])       # Outputs: 0 (default value for int)
print(d1.get('x'))   # Also outputs: 0
```
